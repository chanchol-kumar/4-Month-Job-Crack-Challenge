```python
n = input("Enter a number")
try:
    n = int(n)
    sum = 0
    while n>0:
        sum += n
        n -=1
except ValueError:
    print("Error: Invalid input. Please enter a numeric value.")

print(f'Sum of all number : {sum}')
```
