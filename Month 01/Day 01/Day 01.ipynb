```python
n = input("Enter a number: ")

try:
    num = int(n)
    print(f"You entered the number: {num}")   
except ValueError:
    print("Error: Invalid input. Please enter a numeric value.")
```

```python
x, y = 100, 200
print(f"Before swaping : X = {x}, Y = {y}")
x, y = y, x
print(f"After swaping X = {x}, Y = {y}")
```

```python
n = input("Enter a raw numbers of seconds: ")

try:
    n = int(n)
    hour = n//3600
    min = (n - (hour*3600))//60
    sec = (n - (hour*3600) - (min*60))
    
    print(f"Hours: {hour}")
    print(f"Minutes: {min}")
    print(f"Seconds: {sec}")
except ValueError:
    print("Error: Invalid input. Please enter a numeric value.")
```
